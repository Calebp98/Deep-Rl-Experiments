```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

def meanDiff(run1,run2,run3):
    data0 = np.load(run1)
    data1 = np.load(run2)
    data2 = np.load(run3)
    
    minimum = np.minimum(data0,data1)
    minimum = np.minimum(minimum,data2)

    maximum = np.maximum(data0,data1)
    maximum = np.maximum(maximum,data2)
    
    mean = (data0+data1+data2)/3 
    
    index = np.linspace(1, data0.size , num=data0.size)

    
    return minimum,maximum,mean,index


    
```

```python
Rminimum,Rmaximum,Ravg,Rindex = meanDiff("Rq_1.npy","Rq_2.npy","Rq_3.npy")
Mminimum,Mmaximum,Mavg,Mindex = meanDiff("Rm_1.npy","Rm_2.npy","Rm_3.npy")

# 
```

```python
fig, ax1 = plt.subplots()

ax1.set_xlabel('Game')
ax1.set_ylabel('Reward',color = 'b')
ax1.plot(Ravg, color = 'b',label='Reward')
ax1.fill_between(Rindex,Rminimum,Rmaximum,color='b',alpha=.5)
ax1.tick_params(axis='y', labelcolor='b')

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
ax2.set_ylabel('Moves',color = 'r')
ax2.plot(Mavg,color = 'r',label='Moves')
ax2.fill_between(Mindex,Mminimum,Mmaximum,color='r',alpha=.5)
ax2.tick_params(axis='y', labelcolor='r')



lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc=0)

plt.show()
```
